## 1.2 Calculating totals for subscales and Renaming traits 

```python
#A function that loops in 2 entry values
def score(entry, trait):
    
    '''Function that calculate the Total Just score
    of each of the subscale(Reffering to: conscientiousness,
    emotional stability, openness to new experience,
    agreeableness, extraversion). As set out in the scoring.'''
    
    numbers=0
    _r= re.compile('\d+') #Regex compiles all numbers in entry
    # Treat every entry as two columns
    for col in entry:
        num= _r.findall(col)
        #increment numbers by the trait value it find 
        #at the first index
        if str(trait)==num[0]:
            numbers+= int(num[1])
        else:
            numbers= numbers  
    return numbers

#Changing numbers symbolising traits into factors
factor= dict({
    1 : 'extraversion', 
    2 : 'agreeableness',
    3 : 'conscientiousness', 
    4 : 'emotional stability',
    5 : 'intellect'})

#mapping the score function to the newly
#cretead columns in the data frame
#to sum the totals for every trait atrribute
for i in factor.keys():
    df_person[f'{factor[i]}']= df_person.loc[:,'Section 5 of 6 [I am always prepared.]':'Section 5 of 6 [I worry about things.]'].apply(score, args=(i,) , axis=1)
```

## 2. Identifying High Risk entries

**2.1 Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness**

```python
#df_low: Selecting required columns 
df_low= df_merged.loc[:,['ID','Department','emotional stability',
                           'conscientiousness','agreeableness']]

#df_low_factor: Data frame of scores of certain traits
#with scores of < 30
df_low_factor = df_low.loc[(df_low['emotional stability']<30) &
                           (df_low['conscientiousness']<30) & 
                           (df_low['agreeableness']<30)]
df_low_factor
```

Above is the two entries from the Data Department and Copywriting departments that show to be of high risk based on them having below 30 scores in certain personality traits

**2.2 Print the ID numbers and departments of these applicants to the screen, and also assign these applicants the tag “high_risk” in a new column. All other applicants get the tag “low_risk”**

```python
#F(risk_level): Assesses if an entry is of high or low risk
def risk_level(entry):
    for i in df_merged['ID']:
        if i not in df_low_factor['ID']:
            return 'High Risk'
        else:
            return 'Low Risk'
        
# df_low_factor['Risk Level']:
# Column that shows risk_level of an entry
df_low_factor['Risk Level']= df_low_factor[['emotional stability', 'conscientiousness', 'agreeableness']].apply(risk_level, axis=1)
```
